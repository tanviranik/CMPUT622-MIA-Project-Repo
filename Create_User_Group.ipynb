```python
import pandas as pd
import numpy as np
df = pd.read_csv('uber-raw-data-sep14_processed.csv', encoding = "ISO-8859-1", engine='python')
print(df)
```

```python
size = 6        # sample size
replace = True  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
df2 = df.groupby('MedianIncome', as_index=False).apply(fn)
df2 = df2[['hour', 'lat', 'long', 'base', 'MedianIncomeByPlaceofBirth', 'MedianIncome', 'AvgPopulation' ]]
```

```python
df2
```

```python
def income_level(row):
    if row['MedianIncome'] > 65000 :
        return 1
    else:
        return 0
    
df2['UserGroup'] = df2.apply(lambda row: income_level(row), axis=1)
df2
```

```python
df2.groupby('UserGroup').size()
```

```python
from sklearn.utils import shuffle
df2 = shuffle(df2)
df2.to_csv('uber_census.csv', index=False)
```

```python
def income_level2(row):
    if row['MedianIncome'] < 20000:
        return 0
    elif row['MedianIncome'] >= 20000 and row['MedianIncome'] < 40000:
        return 1
    elif row['MedianIncome'] >= 40000 and row['MedianIncome'] < 60000:
        return 2
    elif row['MedianIncome'] >= 60000 and row['MedianIncome'] < 80000:
        return 3
    else:
        return 4
    
size = 4        # sample size
replace = True  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
df2 = df.groupby('MedianIncome', as_index=False).apply(fn)
df2['UserGroup'] = df2.apply(lambda row: income_level2(row), axis=1)
```

```python
print(df2.groupby('UserGroup').size())
from sklearn.utils import shuffle
df2 = shuffle(df2)
df2.to_csv('uber_census_multiclass.csv', index=False)
```

## Creating another copy of the original dataset

```python
def income_level(row):
    if row['MedianIncome'] > 65000 :
        return 1
    else:
        return 0

import pandas as pd
import numpy as np
df = pd.read_csv('uber-raw-data-sep14_processed.csv', encoding = "ISO-8859-1", engine='python')
print(df)


size = 1       # sample size
replace = True  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
df2 = df.groupby('MedianIncome', as_index=False).apply(fn)
df2
```
